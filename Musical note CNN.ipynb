```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
# BATCH SIZE
# batch size refers to the number of samples (data points or images) that are processed together in each iteration during training.
# batch size is an instance in datloader class
```

```python
#GET ITEM
# The  get_item  in dataloader class use thebatch_size  to determine which subset of data(images,labels) is used to process for each iteration.
# in this case the batch_size  extracts the subset of self.bs  samples from the data to process for iteration.
# smaler bacth   size may slow up the training but  the data will be  genralized more  
# where as for the large batch size the generalisation will be less but  the training will be fast.
```

```python
# The X and Y  empty matrices  crated  in get_ite function accumulate the batch of input images/labels from the batch size
```

```python
#  BATCH PATHS AND BATCH _LABELS
# batch_paths: This variable represents a subset of file paths corresponding to the images in the current batch.
#              It is obtained by slicing the array of all file paths (self.path) from index i to i + self.bs.

# batch_labels: Similarly, this variable represents a subset of labels corresponding to the images in the current batch. 
#               It is obtained by slicing the array of all labels (self.label) from index i to i + self.bs
# These are used to load the image and assign the correct label to the respective position in the batch.

#    In summary, batch_paths and batch_labels are used to extract a subset of file paths and labels for the images in the current batch, 
#     - allowing the data loader to efficiently load and preprocess a batch of data during each iteration.
```

```python
class dataloader(tf.keras.utils.Sequence):
       
    def __init__(self, batch_size, img_width, img_height, data):
        self.bs = batch_size
        self.h = img_height
        self.w = img_width
        self.path = data['path'].values
        self.classname = data['classname'].values
    
    def __len__(self):
        return len(self.path) // self.bs
        
    def __getitem__(self, idx):
        
        i = idx * self.bs
        batch_paths = self.path[i : i + self.bs]
        batch_classname = self.classname[i : i + self.bs]
        
        X = np.zeros((self.bs, self.h, self.w, 3), dtype="float32")
        y = np.zeros((self.bs, 5), dtype="int32")

        for j in range(self.bs): 
            img = load_img(batch_paths[j], color_mode = "rgb", target_size=(self.h, self.w)) # color_mode = "grayscale" 
            img = np.array(img, dtype = 'float32') 
            img = 1-(img/127.5)
            X[j] = img     
            y[j,batch_classname[j]] = 1  
        return X, y
```

```python
train_gen = dataloader(5,224,224,train)
test_gen = dataloader(5,224,224,test)
```

```python
batch1 = test_gen[3]

images = batch1[0]

labels = batch1[1]

print("images in batch = ", images.shape)
print("classname in batch = ", labels.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten

model = Sequential()
model.add(Conv2D(5, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(Conv2D(4, (4, 4), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(4, (3, 3), activation='relu'))
model.add(Conv2D(4, (3, 3), activation='relu'))
model.add(Conv2D(3, (3, 3), activation='relu'))
model.add(Conv2D(3, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss= 'categorical_crossentropy', optimizer ='SGD' , metrics = ['accuracy'])
```

```python
model.fit(train_gen, epochs=30, validation_data= test_gen, verbose=1)
```
